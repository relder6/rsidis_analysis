```python
import uproot
import numpy as np
import matplotlib.pyplot as plt

#Run Number
run_raw = 6126

# File path
raw_data_path = f'/work/hallc/c-rsidis/gaskelld/example_output/coin_replay_production_{run_raw}_-1.root'

# Defining extracted variables
variables = ["T.coin.hRF_tdcMultiplicity","T.coin.hRF_tdcTime", "T.coin.hRF_tdcTimeRaw", "T.coin.pHEL_MPS_adcMultiplicity", "T.coin.pHEL_NEG_adcMultiplicity", "T.coin.pHEL_POS_adcMultiplicity", "T.coin.pRF_tdcMultiplicity", "T.coin.pRF_tdcTime", "T.coin.pRF_tdcTimeRaw", "T.coin.pTRIG1_ROC1_tdcMultiplicity", "T.coin.pTRIG1_ROC1_tdcTime", "T.coin.pTRIG1_ROC1_tdcTimeRaw", "T.coin.pTRIG1_ROC2_tdcMultiplicity", "T.coin.pTRIG1_ROC2_tdcTime", "T.coin.pTRIG1_ROC2_tdcTimeRaw", "T.coin.pTRIG4_ROC1_tdcMultiplicity", "T.coin.pTRIG4_ROC1_tdcTime", "T.coin.pTRIG4_ROC1_tdcTimeRaw", "T.coin.pTRIG4_ROC2_tdcMultiplicity", "T.coin.pTRIG4_ROC2_tdcTime", "T.coin.pTRIG4_ROC2_tdcTimeRaw"]

# Function to load data from ROOT files
def load_data(file_path):
    file = uproot.open(file_path)
    tree = file['T']
    return tree.arrays(variables, library="np")

# Load raw and dummy data
data_raw = load_data(raw_data_path)


# Define histogram settings
varname = "T.coin.hRF_tdcMultiplicity"
var = data_raw[varname]
bin_min, bin_max, bin_number = 0, 5, 100
bins = np.linspace(bin_min, bin_max, bin_number + 1)

# Plot raw data histogram
plt.hist(var, bins=bins, histtype='step', color='red', label=f'Raw Data Run {run_raw}', alpha=0.5, linewidth=2, log=True)
plt.xlabel(varname)
plt.ylabel("Counts")
plt.show()
```
